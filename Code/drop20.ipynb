```python
import numpy as np
import pandas as pd
import random
```

```python
df_ratings  = pd.read_csv('C:/Users/Sunwoo/Desktop/Matrix Factorization/ml-latest-small/ml-latest-small/ratings.csv')
```

```python
df_ratings
```

```python
df_ratings = df_ratings.drop(columns=['timestamp'])
df_ratings = df_ratings.reset_index()
```

```python
df_ratings
```

```python
# 전체 데이터의 20%만 남기기 위해 인덱스 난수
# targets = [random.randint(0,100836) for _ in range(100836//5)]
targets=random.sample(range(0,100836),(100836//5))
targets.sort()
targets
```

```python
dropped= []
for i in range(len(targets)):
    dropped.append(tuple(df_ratings.loc[targets[i]]))

# dropped = df_ratings.loc[[targets[i] for i in range(len(targets))]
# dropped = df_ratings.loc[~df_ratings[['index'].isin(targets)]
# dropped = pd.DataFrame(dropped)
dropped = pd.DataFrame(dropped, columns = ['index','userId','movieId','rating'])
dropped
```

```python
# dropped = dropped.reset_index(['index'])
# dropped[['index','userId','movieId']] = dropped[['index','userId','movieId']].apply(pd.to_numeric)
dropped = dropped.astype(int)
dropped = dropped.astype({'rating':np.float})
dropped = dropped.drop(columns=['index'])
dropped
```

```python
rating_80 = df_ratings.loc[~df_ratings['index'].isin(targets)]
rating_80 = rating_80.drop(columns=['index'])
rating_80
```

```python
dropped.to_csv('C:/Users/Sunwoo/Desktop/Matrix Factorization/result/dropped.csv')
rating_80.to_csv('C:/Users/Sunwoo/Desktop/Matrix Factorization/result/rating_80.csv')
```
